### **5 Control Flow in R**
Control flow structures allow you to control the execution of code based on conditions or to repeat tasks. These are essential for writing dynamic and flexible programs.

#### **5.1 If-Else Statements**
Use if and else to execute code based on conditions.

````r
# Example 1: Simple if-else
x <- 10

if (x > 5) {
  print("x is greater than 5")
} else {
  print("x is 5 or less")
}

# Example 2: if-else ladder
score <- 85

if (score >= 90) {
  print("Grade: A")
} else if (score >= 80) {
  print("Grade: B")
} else {
  print("Grade: C")
}
````

#### **5.2 For Loops**
For loops iterate over a sequence (e.g., vector, list).

````r
# Example: Loop through a vector
numbers <- c(1, 2, 3, 4, 5)

for (num in numbers) {
  print(num^2)  # Print the square of each number
}
````

#### **5.3 While Loops**
While loops repeat code while a condition is TRUE.

````r
# Example: While loop
counter <- 1

while (counter <= 5) {
  print(counter)
  counter <- counter + 1  # Increment the counter
}
````

#### **5.4 Repeat Loops**
Repeat loops continue until explicitly stopped with break.

````r
# Example: Repeat loop
x <- 1

repeat {
  print(x)
  x <- x + 1

  if (x > 5) {
    break  # Exit the loop
  }
}
````

#### **4.5 Using the `apply` Family**
The `apply` family of functions provides efficient alternatives to loops for repetitive operations.

* `apply`: Apply a function to rows or columns of a matrix.
* `lapply`: Apply a function to each element of a list.
* `sapply`: Similar to lapply, but simplifies the output.

````r
# Example: apply
mat <- matrix(1:9, nrow = 3)

# Sum of each row
row_sums <- apply(mat, 1, sum)
print(row_sums)

# Example: lapply and sapply
list_data <- list(a = 1:5, b = 6:10)

# Find the mean of each element
lapply(list_data, mean)  # List output
sapply(list_data, mean)  # Vector output
````

### **6 Functions in R**
Functions are reusable blocks of code that perform specific tasks. In R, you can use built-in functions or create your own. Let’s explore both.

#### **6.1 Writing Custom Functions**
To define a function, use the `function()` keyword.

````r
# Example: A simple function
add_numbers <- function(a, b) {
  sum <- a + b
  return(sum)  # Return the result
}

# Call the function
result <- add_numbers(5, 10)
print(result)  # Output: 15
````

#### **6.2 Arguments and Default Values**
Functions can have arguments with default values.

````r
# Function with default arguments
greet <- function(name = "User") {
  message <- paste("Hello,", name)
  return(message)
}

print(greet())            # Uses default: "User"
print(greet("Alice"))     # Output: "Hello, Alice"
````

#### **6.3 Scoping Rules**
Variables defined inside a function are local to that function.

````r
# Example of scope
example_function <- function() {
  x <- 10  # Local variable
  return(x)
}

print(example_function())  # Output: 10
# print(x)  # Error: x not found (x is local to the function)
````

#### **6.4 Anonymous Functions**
Functions can be defined without a name, often used in quick one-off tasks.

````r
# Example: Anonymous function
result <- (function(a, b) a * b)(5, 3)
print(result)  # Output: 15
````

#### **6.5 Common Built-in Functions**
R includes many built-in functions for data analysis, such as:

1. `mean()`: Calculate the average.
2. `sd()`: Calculate the standard deviation.
3. `sort()`: Sort a vector.
4. `paste()`: Concatenate strings.

````r
# Example: Built-in functions
numbers <- c(1, 2, 3, 4, 5)

print(mean(numbers))  # Average
print(sd(numbers))    # Standard deviation
print(sort(numbers, decreasing = TRUE))  # Descending order
````

### **7 Data Manipulation with dplyr**
The dplyr package, part of the tidyverse, is a powerful tool for data manipulation in R. It provides simple and intuitive functions to filter, transform, and summarize data.

#### **7.1 Installing and Loading dplyr**
Install the tidyverse package if you don’t have it, which includes dplyr:

````r
# Install the package
install.packages("tidyverse")

# Load the package
library(dplyr)
````

#### **7.2 The dplyr Workflow**
The core functions in dplyr focus on common data manipulation tasks. Here are the most important ones:

1. `filter()`: Subset rows based on conditions.
2. `select()`: Choose specific columns.
3. `mutate()`: Add or modify columns.
4. `arrange()`: Sort rows.
5. `summarize()` and `group_by()`: Aggregate data.

Let’s explore each function.

#### **7.3 Filter Rows with `filter()`**

````r
# Example data frame
df <- data.frame(
  Name = c("Alice", "Bob", "Charlie", "Diana"),
  Age = c(25, 30, 35, 40),
  Score = c(90, 85, 88, 95)
)

# Filter rows where Age > 30
filtered_df <- filter(df, Age > 30)
print(filtered_df)
````

#### **7.4 Select Columns with `select()`**

````r
# Select only the Name and Score columns
selected_df <- select(df, Name, Score)
print(selected_df)
````

#### **7.5 Add or Modify Columns with `mutate()`**

````r
# Add a new column for scores squared
mutated_df <- mutate(df, Score_Squared = Score^2)
print(mutated_df)
````

#### **7.6 Sort Rows with `arrange()`**

````r
# Arrange rows by Score in descending order
sorted_df <- arrange(df, desc(Score))
print(sorted_df)
````

#### **7.7 Aggregate Data with `summarize()` and `group_by()`**

````r
# Example grouped summary
grouped_summary <- df %>%
  group_by(Age > 30) %>%  # Group by whether Age is greater than 30
  summarize(Avg_Score = mean(Score), Count = n())

print(grouped_summary)
````

#### **7.8 The Pipe Operator `%>%`**
The `%>%` operator allows chaining operations for a clean and readable workflow.

````r
# Chain multiple operations
result <- df %>%
  filter(Age > 30) %>%
  select(Name, Score) %>%
  arrange(desc(Score))

print(result)
````

### **8 Data Visualization with ggplot2**
The ggplot2 package, part of the tidyverse, is one of the most popular tools for data visualization in R. It uses the "grammar of graphics" to create layered, customizable plots.

#### **8.1 Installing and Loading ggplot2**
If you already installed tidyverse, you have ggplot2. To load it:

````r
install.packages("tidyverse")
library(ggplot2)
````

#### **8.2 Understanding the Grammar of Graphics**
The main idea behind ggplot2 is to build a plot layer by layer. The components are:

- Data: The dataset you want to visualize.
- Aesthetics (`aes()`): Map variables to visual properties like x, y, color, etc.
- Geometries (`geom_*`): The type of plot, such as points, lines, bars, etc.
- Facets: Create subplots based on categories.
- Themes: Adjust non-data elements (backgrounds, grids, etc.).

#### **8.3 Creating a Basic Plot**
Let’s use the mtcars dataset, which comes built-in with R.

````r
# Scatterplot: mpg vs. wt
ggplot(data = mtcars, aes(x = wt, y = mpg)) +
  geom_point()
````

#### **8.4 Adding Layers to Enhance the Plot**
You can add layers to make your plot more informative:

````r
# Scatterplot with colors and sizes
ggplot(data = mtcars, aes(x = wt, y = mpg, color = cyl, size = hp)) +
  geom_point() +
  labs(
    title = "Miles per Gallon vs. Weight",
    x = "Weight (1000 lbs)",
    y = "Miles per Gallon",
    color = "Cylinders",
    size = "Horsepower"
  ) +
  theme_minimal()
````

#### **8.5 Creating Other Common Plots**
- Bar Plot

````r
# Count of cars by number of cylinders
ggplot(data = mtcars, aes(x = factor(cyl))) +
  geom_bar() +
  labs(x = "Cylinders", y = "Count")
````

- Histogram

````r
# Distribution of mpg
ggplot(data = mtcars, aes(x = mpg)) +
  geom_histogram(binwidth = 5, fill = "blue", color = "black") +
  labs(x = "Miles per Gallon", y = "Frequency")
````

- Line Plot

````r
# Line plot of mpg vs. wt
ggplot(data = mtcars, aes(x = wt, y = mpg)) +
  geom_line() +
  labs(title = "Miles per Gallon vs. Weight")
````

#### **8.6 Facets for Subplots**
Facets split your plot into multiple subplots based on a categorical variable.

````r
# Facet by number of cylinders
ggplot(data = mtcars, aes(x = wt, y = mpg)) +
  geom_point() +
  facet_wrap(~ cyl) +
  labs(title = "Miles per Gallon vs. Weight by Cylinders")
````

#### **8.7 Customizing with Themes**
Themes control the appearance of non-data elements. Use built-in themes or create your own.

````r
# Applying a theme
ggplot(data = mtcars, aes(x = wt, y = mpg)) +
  geom_point() +
  theme_classic()  # Classic theme
````

### **10 Statistical Analysis in R**
In this step, we'll explore how to perform basic statistical analysis using R. R is a powerful tool for both descriptive and inferential statistics, making it suitable for data analysis and research.

#### **10.1 Descriptive Statistics**
Descriptive statistics summarize and describe the characteristics of a dataset.

- **Mean**: The average value of a variable.

````r
# Mean of a column
mean(data$median_house_value, na.rm = TRUE)
````

- **Median**: The middle value when the data is ordered.

````r
# Median of a column
median(data$median_house_value, na.rm = TRUE)
````

- **Standard Deviation**: A measure of how spread out the values are.

````r
# Standard deviation of a column
sd(data$median_house_value, na.rm = TRUE)
````

- **Summary Statistics**: A quick summary of key statistics.

````r
# Summary of a column or dataset
summary(data$median_house_value)
summary(data)
````

- **Quantiles**: Values that divide the data into equal parts.

````r
# Quantiles of a column
quantile(data$median_house_value, probs = c(0.25, 0.5, 0.75))
````

#### **10.2 Hypothesis Testing**
Hypothesis tests are used to make inferences about population parameters based on sample data.

- **T-Test**: A t-test compares the means of two groups.

````r
# # One-sample t-test (mean of a sample vs a known value)
# t.test(data$median_house_value, mu = 50)  # Test if the mean is 50

# # Two-sample t-test (compare means of two groups)
# t.test(data$median_house_value ~ data$median_income)
````

- **Chi-Square Test**: Used to test the association between categorical variables.

````r
# Chi-square test of independence
table_data <- table(data$median_house_value, data$median_income)
chisq.test(table_data)
````

- **ANOVA**: Compares the means of more than two groups.

````r
# One-way ANOVA
anova_result <- aov(median_house_value ~ median_income, data = data)
summary(anova_result)
````

#### **10.3 Correlation and Regression Analysis**
- **Correlation**: Measures the strength and direction of the linear relationship between two variables.

````r
# Correlation between two variables
cor(data$median_house_value, data$median_income, use = "complete.obs")
````

- **Simple Linear Regression**: Models the relationship between two variables.

````r
# Fit a linear model
model <- lm(median_house_value ~ median_income, data = data)

# Summary of the model
summary(model)
````

- **Multiple Linear Regression**: Models the relationship between multiple predictors and a response variable.

Confidence intervals provide a range of values that are likely to contain the true population parameter.
